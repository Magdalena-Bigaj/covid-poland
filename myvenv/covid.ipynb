```python
csv_data= pd.read_csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv")
print(type(csv_data))
```

```python
csv_data['dateRep'] = pd.to_datetime(csv_data['dateRep'],format='%d/%m/%Y')
csv_data.head(30)
```

```python
csv_data.sort_values('dateRep')
```

```python
csv_data_poland = csv_data.loc[csv_data['countriesAndTerritories'] == 'Poland']
csv_data_poland.head()
```

```python
fig, ax_left = plt.subplots(figsize=(20, 8))

ax_right = ax_left.twinx()
plt.plot(csv_data_poland['dateRep'], csv_data_poland['cases'], color= '#003049', label= 'DAILY CASES')
plt.plot(csv_data_poland['dateRep'], csv_data_poland['deaths'], color= '#d62828', label= 'DAILY DEATHS')
plt.tight_layout
plt.legend
```
